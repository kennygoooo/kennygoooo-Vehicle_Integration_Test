```python
# Python program 
# To print numbers 1-250 for divisibility by 3, 5, and 7

for i in range(1, 251):
    output = ""
    
    if i % 3 == 0:
        output += "Fizz"
    if i % 5 == 0:
        output += "Fizzy"
    if i % 7 == 0:
        output += "Buzz"
    
    if output == "":
        print(i)
    else:
        print(output)
```
